```python
# run_dataloader.py
from pathlib import Path
from data_loader import DataLoader  # make sure data_loader.py is in the same folder

def main():
    # --- paths (adjust to your file locations) ---
    responses_path = Path("./data/M250512 Budget Thuis Multi utility onderzoek MaxDiff_data_volledig_opgeschoond.csv")   # Dataset 1
    design_path = Path('./data/M250512_MXD_Design.xlsx')        # Dataset 2
    output_path = Path("maxdiff_compact.csv")

    # --- run pipeline ---
    dl = DataLoader()
    responses = dl.load_responses(responses_path)
    design = dl.load_design(design_path)
    compact = dl.combine(responses, design)
    dl.write_csv(compact, output_path)

    print("\nFirst rows of compact table:")
    print(compact.head())

if __name__ == "__main__":
    main()
```
